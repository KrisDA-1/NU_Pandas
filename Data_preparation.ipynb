DATA CLEANING

-Missing values


```python
##Libraries
import pandas as pd
import numpy as np

##read data
df=pd.read_csv('Iris_Modified.csv')

##set seed for reproducibility
np.random.seed(0)
```

```python
##Data exploration
df.sample(10)
```

```python
##Missing data points per column
mv_count = df.isnull().sum()
```

```python
##Count of missing points in the first ten columns
mv_count[0:10]
```

```python
##Total missing values
t_cells=np.prod(df.shape)
t_missing=mv_count.sum()
```

```python
##Percent of missing data
(t_missing/t_cells)*100
```

```python
##Remove all the rows that contain a missing value
df.dropna()
```

```python
##Remove all columns with at least one missing value
columns_with_na_dropped=df.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
##Counting data we lost
print("Columns in original dataset: %d \n" % df.shape[1])
print("Columns with na's dropped: %d" % columns_with_na_dropped.shape[1])
```

-Filling in missing values automatically

```python
##Get a small subset from dataset
subset_df=df.loc[:, 'SepalLengthCm':'PetalWidthCm'].head()
display(subset_df.head())
```

```python
##Replace all Na with 0
subset_df.fillna(0)
```

```python
##Replace all NA's the value that comes directly after it in the same column
##Then replace all the remaining na's with 0
subset_df.fillna(method='bfill', axis=0).fillna(0)
```
